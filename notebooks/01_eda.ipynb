# 01 — Exploratory Data Analysis

Quick look at the Pima Indians Diabetes dataset before any modelling.
The source-of-truth pipeline lives in `src/diabetes_mlops/` — this notebook is for exploration only.

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent / 'src'))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from diabetes_mlops.data import load_raw, clean
from diabetes_mlops import config as C

sns.set_theme(style='whitegrid')
```

```python
raw = load_raw()
print(raw.shape)
raw.head()
```

## Domain note — zero-as-missing

`Glucose`, `BloodPressure`, `SkinThickness`, `Insulin`, and `BMI` cannot be zero in a living patient. These zeros encode missing values.

```python
raw[C.ZERO_IS_MISSING].eq(0).sum().sort_values(ascending=False)
```

```python
df = clean(raw)
print('class balance:', df[C.TARGET].value_counts(normalize=True).round(2).to_dict())
df.describe().T
```

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 7))
for ax, col in zip(axes.flat, C.FEATURES):
    sns.kdeplot(data=df, x=col, hue=C.TARGET, common_norm=False, fill=True, ax=ax)
    ax.set_title(col)
plt.tight_layout()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Pairwise correlations')
plt.tight_layout()
```
